# Item Types and counts

```python
shop_c=pd.DataFrame({'Item':shop.Item_Type.value_counts().index,'Count':shop.Item_Type.value_counts().values,})
shop_c
```

# Analysing the Item price List

```python
plt.figure(figsize=(7,8),dpi=200)
ax=sns.barplot(y='Item',x='Count',data=shop_c)
for i in ax.containers:
    ax.bar_label(i,fmt='%.1f')
plt.show()
```
